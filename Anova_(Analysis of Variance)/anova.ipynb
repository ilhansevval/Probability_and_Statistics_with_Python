# **ANOVA (Analysis of Variance)**

Z Distribution ve T Distribution ile 2 ya da daha fazla sub group'un aynı population'dan olma olasılığına bakılır.

F distribution ile ise aynı variance'a sahip 2 ya da daha fazla sub group'un aynı population'dan olma olasılığına bakılır.

ANOVA ile gruplar incelendiğinde aynı population'ın farklı grupları olup olmadığı anlaşılmış olur.

ANOVA iki çeşit variance inceler;

**Gruplar arasındaki variance:** Grupların mean'inin total mean'den ne kadar uzak olduğudur.

**Gruplar içerisindeki variance:** Gruplardaki değerlerin ait olduğu grubun mean'in ne kadar uzak olduğudur.

F değeri = (Gruplar arasındaki variance) / (Gruplar içerisindeki variance)

Örnek olarak kırsal bir bölgede yaşayan insanların stres düzeyleri ile kentsel bir bölgede yaşayan insanların stres düzeyleri karşılaştırılacaktır.

**Null Hypothesis:** Kırsal bölgede ve kentsel bölgede yaşayan insanların stres düzeyleri aynıdır.

reject durumu stres düzeylerinin aynı olmadığı durumdur.

fail to reject durumu stres düzeylerinin aynı olduğu durumdur.

```python
import numpy as np
kırsal = np.array([45.93, 34.71, 78.96, 42.19, 91.56,
                   37.98, 76.41, 36.15, 88.41, 39.84])
kentsel = np.array([85.78, 93.72, 25.16, 28.16, 17.13,
                    68.93, 23.11, 69.32, 79.17, 93.12])
```

```python
print("kırsal bölge mean: ",np.mean(kırsal))
print("kentsel bölge mean: ",np.mean(kentsel))
total_mean = (np.mean(kırsal) + np.mean(kentsel))/2
print("total mean: ",np.mean(total_mean))
```

Grup ortalamaları ve total ortalamaya bakılmıştır.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(kırsal)
sns.kdeplot(kentsel)
plt.show()
```

F distribution'da sola yatıklık vardır, pozitif skewnesslık olduğu görülür.



```python
from scipy import stats
stats.f_oneway(kırsal, kentsel)
```

static sonucu F değeri'dir.

F değeri eğer;

kritik eşikten büyükse reject durumu vardır.

kritik eşikten küçükse fail to reject durumu vardır.

F tablosunda **degrees of freedom for groups** ve **degrees of freedom for error** değeri vardır. 

degrees of freedom for groups değeri, grup sayısı - 1 ile bulunur.

Örnek için degrees of freedom for groups değeri 2-1=1'dir.

degrees of freedom for error değeri, (gruplardaki sample sayısı - 1) * grup sayısı ile bulunur.

Örnek için degrees of freedom for error değeri (10-1)*2=18'dir.

F tablosuna bakıldığında;

satırlar degrees of freedom for error değeridir. Örnek için 18 bulunmuştu.

sütunlar degrees of freedom for groups değeridir. Örnek için 1 bulunmuştu.

image.png

F tablosundaki confidence level %99'dur.

100 testin 99'unda null hypothesis reject edilmiştir.

Örnek için F tablosundaki kritik değer 4.41'dir.

F değeri ise 0.008'dir. 

4.41 > 0.008'den olduğu için fail to reject durumu gerçekleşmiştir.

Yapılan gözlemlere göre kırsal bölgede ve kentsel bölgede yaşayan insanların stres düzeyleri aynıdır sonucuna ulaşılmıştır.
