# **Permutation**

Her sembolün sadece bir veya birkaç kez kullanıldığı sıralı bir dizidir.

c2a86ff92e4d76bea51ff281b5867c2f-300x101.gif

5 kişinin 3 kişilik bir banka kaç farklı şekilde oturacağı 5'in 3'lü permütasyonu ile bulunur.



```python
#P(5,3)
import math
pay = math.factorial(5)
payda = math.factorial(2)
P = int(pay / payda)
print(P)
```

3 farklı pantolon, 4 farklı ceket ve 2 farklı gömleği olan kişi bir ceket bir gömlek ve bir pantolonu kaç farklı şekilde giyebileceğine permütasyon hesabı ile bakılırsa;

```python
#P(3,1)
import math
pay = math.factorial(3)
payda = math.factorial(2)
pantolon = int(pay / payda)
print("pantolon secimi: ",pantolon)
```

```python
#P(4,1)
import math
pay = math.factorial(4)
payda = math.factorial(3)
ceket = int(pay / payda)
print("ceket secimi: ",ceket)
```

```python
#P(2,1)
import math
pay = math.factorial(2)
payda = math.factorial(1)
gomlek = int(pay / payda)
print("gomlek secimi",gomlek)
```

```python
permutasyon = pantolon*ceket*gomlek
print("permutasyon sonucu: ",permutasyon)
```

# **Independent and Dependent Events**

**Independent Event:** İki olayın birbirinden bağımsız olma durumudur.

Havaya para atıldığında, daha önce gelen yazı-tura sonuçlarının o an havaya atılan paraya herhangi bir etkisi yoktur.

Bu durum independent event'e bir örnektir.

**Dependent Event:** İki olayın gerçekleşmesinin birbirine bağlı olduğu durumdur.

Bir torba sarı,mavi bilyeler olsun.

Mavi ve sarı bilyelerin seçilme olasılığı birbirne bağlıdır.

Çünkü her seçimde bir renk azalacağından seçilmeyen rengin olasılığı artacaktır.

Başka bir örnekle anlatılacak olursa;

Bir datada yaş ve maaş featurelarından oluşsun.

Bu dataya bakıldığından yaş arttıkça terfi alındığından maaşın da arttığı görülecektir.

Bu iki feature dependent featurelardır.

Bir datada birden çok feature birbiriyle ilişkili olduğunda (multicollinearity) ve aralarındaki ilişki de yüksekse bu feature'ların hepsini kullanmak modeli yanıltacaktır.

Bu yüzden en önemli feature seçilerek model kurulur.

# **Conditional Probability**

Bir olayın gerçekleşme olasılığı başka bir olayın gerçekleşme olasılığına bağlıdır.

Yani ilk olay gerçekleştiği takdirde ikinci olay gerçekleşebilir.

İki torba olsun. A torbasında 3 mavi, 5 siyah top, B torbasında 2 mavi 3 siyah top olsun. Torbalardan birinden alınan topun mavi olduğu bilindiğine göre bu topun B torbasından olma olasılığına bakıldığında,

P(B|M) = P (B and M) / P(M) = (2/3) / ((3/8) + (2/5)) = 80/93 'tür.

# **Bayes Theorem**

Bir olasılık değerinin bilinen diğer olasılıkları kullanarak hesaplama fikridir.

1_LEvKQxfhsAfa47YwNflF2g.png

**P(A|B)** : B koşulu gerçekleştiğinde A'nın gerçekleşme olasılığı

**P(B|A)** : A koşulu gerçekleştiğinde B'nın gerçekleşme olasılığı

Piknik yapmayı düşünen bir grup havanın bulutlu olduğunu görmüştür. Yağmurlu günlerin %40'ının bulutlu bir günle başladığı bilinmektedir ve günlerin %40'ı bulutlu başlamaktadır. Bir günün yağmurlu gün olma olasılığı %10 olduğuna göre bulutlu başlanan bir günde yağmur yağma olasılığı,

P(Yağmurlu|Bulutlu) = (P(Bulutlu|Yağmurlu)*P(Yağmurlu)) / P(Bulutlu)

P(Bulutlu|Yağmurlu) = 0.50

P(Yağmurlu) = 0.10

P(Bulutlu) = 0.40

(0.50*0.10) / 0.40 = 0.125 

Bulutlu başlanan bir günde yağmur yağma olasılığı 0.125'tir.
