Örneğin bir zar yüzlerce, binlerce kez havaya atıldığında ortaya çıkacak olasılıkların dağılımıdır. Ortaya çıkabilecek olasılıklar bir dağılım halinde bilinir.

**Discrete** ve **Continious** olmak üzere iki farklı dağılım bulunmaktadır.

Discrete kavramı bir örnekle anlatılacak olursa;

Bir zarın para havaya 1 ya da 2 kez atılabilir, 1.13 kez atamayız. Paranın havaya atılması discrete bir olaydır.

Discrete Probability Distributions;

Continuous Probability Distributions;

**Uniform Distributions**

Discrete distribution anlatmak ve anlamak için kullanılan en kolay yöntemlerden biridir.

Bir zarın havaya atılması discrete bir olaydır. Çünkü 6 tane olası sonuç vardır.

Uniform bir dağılımdır, çünkü 6 sonucun çıkma olasılığı da aynıdır.

Uniform Distribution tüm sonuçların aynı olasılıkta çıkmasıdır.

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.random.randint(1,7,60000)
#(inclusive,inclusive,60000 değer)
print("sample space: ",np.unique(a))
plt.figure(figsize = (5,5))
plt.hist(a,bins=12,color="green")
plt.ylabel("Number of outcomes")
plt.xlabel("Possible outcomes")
plt.show()
```

60000 kez zar atıldığında uniform bir dağılım olduğundan her bir sonuçtan 10000 değeri elde edilmesi beklenir.

**Binomial Distributions**

Binomial 2 tane discrete sonucu olan bir trial'dır.

Bernoulli Trial binomial distribution hesaplanırken kullanılır.

Bernoulli Trial success ve failure olmak üzere 2 sonuç vardır.

Bir deneyin birden çok sonucu olabilir. Bu sonuçların bir tanesi success diğerleri de failure olarak tanımlanır ya da bir tanesi failure diğerleri success olarak tanımlanır.



bernoille.jpg

Bir zar atıldığında, zarın sonucu 1 çıkarsa success, geriye kalan 2,3,4,5,6 sonuçları çıkarsa failure olarak tanımlansın.

**n:** zarın kaç defa atıldığı, trial sayısı

**p:** 1 gelme olasılığı 1/6'dır.

**r:** başarılı sonuç sayısı

```python
n = 2
p = 1/2
s = np.random.binomial(n, p,10000) 
weights = np.ones_like(s)/float(len(s))

plt.hist(s, weights=weights,color="purple")
plt.xlabel("number of success") 
plt.ylabel("probability")
plt.show()
```

Madeni paranın 2 kez havaya atılma örneğinde;

2 kez üst üste yazı gelme olasılığı 0.25'tir.

1 kez yazı 1 kez tura gelme olasılığı 0.50'dir. (yazı-tura,tura-yazı)

2 kez üst üste tura gelme olasılığı 0.25'tir.

Plotta 3 output olduğundan 0,1,2 değerlerine sonuçlar yazdırılmıştır.

**Poisson Distributions**

Binomial Distribution'da trial'a bağlı başarıya bakılırken, poisson distribution'da interval'a bağlı başarıya bakılır.

interval zamana bağlı ya da mesafeye bağlı durumdur.

**lambda:** belirli bir zaman aralığında ya da distance'ta istenilen olayın olma olasılığıdır.

posss.jpg

```python
lamda = 10
s1 = np.random.poisson(lamda,100000)
weights1 = np.ones_like(s1)/float(len(s1))
plt.hist(s1, weights=weights1,bins = 100,color="orange")
plt.xlabel("number of occurances") 
plt.ylabel("probability")
plt.show()
```

Bir saatte 10 kamyonun otoparka gelme olasılığına bakılmıştır.

Çevre koşulların ya da başka sebeplerden ötürü her zaman saatte 10 kamyon otoparka gelmez.

Bu 10 kamyon 100000 kez yollandığında elde edilen sonuçlar görselleştirilmiştir.

Plota bakıldığında 10'un çevresinde bir dağılım gerçekleştirilmiştir.

Düşük olasılıklarla da olsa saatte 5 kamyon ya da 15 kamyon gittiği gözlemlenmiştir.

# **Gaussian (Normal) Distributions**

Normal distribution'da mean, mod, median değerleri birbirine eşittir.

Eğrinin maksimum noktası mean değeridir ve verilerin çoğu bu değerin çevresinde kümelenir.

Dağılıma ait değerlerin yaklaşık %68,2'si -1 ve 1 standart sapma aralığında,

Dağılıma ait değerlerin yaklaşık %95,4'ü -2 ve 2 standart sapma aralığında,

Dağılıma ait değerlerin yaklaşık %99,7'si -3 ve 3 standart sapma aralığında yer almaktadır.

-3 ve 3 standart sapma aralığında olmayan değerler outlierdır.

Çok bilindik bir örnekle açıklanılacak olunursa,

IQ testinde ortalama sonucun 100 çıktığını varsayalım.

Süper zekaların sonucu 140-150 çıksa da bu nadir görülen bir durumdur.

Bazı insanların IQ'ları da ortalamadan düşük olarak 40-50 aralığında çıksa da bu da nadir görülen bir durumdur.

Genel dağılım ortalama etrafında toplanır.

```python
mu, sigma = 100, 20  
#mu=mean 
#sigma=standard deviation
s = np.random.normal(mu, sigma, 100000)
print("mean: ", np.mean(s))
print("standart deviation: ", np.std(s))
plt.figure(figsize = (5,5))
plt.hist(s, 100,color="red")
plt.ylabel("frequency")
plt.xlabel("IQ")
plt.title("Histogram of IQ")
plt.show()
```

-1 ve 1 standard deviation aralığında yani 80 ve 120 IQ değerine sahip insanlar tüm insanların %68.2'sini oluşturur.

Böylelikle standard deviation'a bakılarak insanların ortalamaya göre ne kadar yayıldıkları bilgisine ulaşılır.

Değerlerin -3 ve 3 standard deviation aralığında toplandığı görülür.


**Z Score**

Z tablosu (standart dağılım tablosu) ile belirli bir z değeri altında kalan alan bulunur.

z = (x-mean) / std

70 ile 130 arasında IQ'ya sahip insanların yüzdesi bulunmak istenilirse,

z1 = (70-100) / 20 = -1.5

z2 = (130-100) / 20 = 1.5

z tablosuna bakıldığında 1.5 değerinin 0.43 değerini gösterdiği görülür.

Buradan 70 ile 130 IQ'ya sahip insanlar tüm insanların %86'sını kapladığı çıkarımına varılır.
