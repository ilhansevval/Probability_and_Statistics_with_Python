# **Level of Measurement**

4 çeşit Level of Measurement vardır. Bunlar **nominal**, **ordinal**, **interval** ve **ratio**'dur.

**Nominal:** Data **kategoriktir**. Bu yüzden veriler sıralanamazlar.

**Ordinal:** Data **kategoriktir** ve **sıralanabilir** ancak aralarındaki sayısal olarak bir ilişki bulunamaz. 

Mesela bir anket düzenlendiğinde her zaman, sıklıkla, bazen, nadiren sonuçları elde edilir. Her zaman seçeneği bazenden daha genel bir ifadedir. Ancak sayısal olarak ne kadar olduğunun bir karşılığı yoktur. 'Akşamları genelde çay içerim' cümlesi ile 'Akşamları bazen çay içerim' genel ve bazen kelimelerinin categorical olduğu ve geneldenin bazenden büyük olduğu anlaşılır ancak bir ayda ne kadar çay içtiğinin bilgisi bu cümleden çıkarılamaz, bir yüzdesi yoktur. Bu tarz veriler ordinal verilere örnektir.

Nominal ve Ordinal verinin arasındaki fark sıralanabilirliktir.

**Interval**: Datadaki farklılıklar anlamlıdır, belirli bir ölçeği vardır ancak natural zero starting point yoktur ve oranlar anlamsızdır.

Mesela 25 derece 15 dereceden sıcaktır gibi bir karşılaştırma yapılabilir. Ancak odanın sıcaklığı 0 dereceden başlıyor ya da 15 dereceden başlıyor gibi bir çıkarımda bulunulamaz. 

**Ratio:** Natural zero starting point vardır ve oranlar anlamlıdır. Belirli bir ölçeği vardır, sıralanabilirler.

Mesela bir insanın yaşı 0'dan başlar. 

40 km 20 km'den iki kat uzaktır gibi bir çıkarımda bulunabilir.

# **Dispersion**

```python
import matplotlib.pyplot as plt
y = np.random.uniform(5,8,175)
x1 = np.random.uniform(10,20,175)
x2 = np.random.uniform(0,30,175)
plt.scatter(x1,y,color="red")
plt.scatter(x2,y,color="purple")
plt.grid()
plt.xlim([-1,31])
plt.ylim([0,10])
plt.xlabel("x")
plt.ylabel("y")
print("X1 mean: {} and median: {}".format(np.mean(x1),np.median(x1)))
print("X2 mean: {} and median: {}".format(np.mean(x2),np.median(x2)))
```

x1 ve x2 için mean ve median değerlerinin birbirine çok yakın olduğu görülür.

Eğer plota bakılmasaydı x1 ve x2'nin aşağı yukarı aynı data olduğu söylenebilirdi. 

Ancak plota bakıldığında iki datanın birbirinden oldukça farklı olduğu görülür.

Dağılımı ölçmenin, ortalamadan ya da merkezden ne kadar uzakta olduğunu belirtmenin farklı yolları vardır.

**Range:** Bir listedeki en büyük değer ile en küçük değer arasındaki farktır.

```python
list1 = [-10,0,10,20,30]
list2 = [8,9,10,11,12]
```

```python
print("list1 mean:",np.mean(list1))
print("list2 mean:",np.mean(list2))
```

```python
print("Range of list1: ", (np.max(list1)-np.min(list1)))
print("Range of list2: ", (np.max(list2)-np.min(list2)))
```

Ortalamalarına bakıldığında iki listenin de ortalaması aynıdır.

Ancak range'e bakıldığında list1'in range'inin daha fazla olduğu görülür.

Bu durum list1'in daha dağınık olduğunu gösterir.

Aynı range'e sahip olan iki veri olduğunda range'e bakmak yeterli olmayacaktır.

Ancak bu iki verinin dağılımları, birimlerin yerlerine ya da nasıl gruplandıklarına bağlı olarak değişiklik gösterir.

Bu durumda en sık kullanılan yöntemler variance ve standard deviation'dır.

**Variance:** Verilerin ortalamaya olan uzaklarının karelerinin toplamının ortalamasıdır.

Bir verideki gözlemlerin ne kadar yayıldığını gösterir.

Veri setinde varyans küçük olduğunda gözlemlerin yakınlığını gösterirken, büyük olduğunda gözlemlerin birbirlerinden dağılmış halde bulunduğunu gösterir.

varyans-formul.png

```python
print("Variance of list1: ", (np.var(list1)))
var1 = sum((list1 - np.mean(list1))**2)/len(list1)
print("Variance with formula: ",var1)
print("Variance of list2: ", (np.var(list2)))
var2 = sum((list2 - np.mean(list2))**2)/len(list2)
print("Variance with formula: ",var2)
```

Liste1'de gözlemler 10'ar 10'ar artmıştır. Birbirinden uzaktır.

List1'in varyansı 200'dür. Yüksek varyans değeri verinin dağınık halde bulunduğunu gösterir.

Eğer list1'deki tüm değerler bilinmeseydi varyansa bakarak verilerin birbirinden uzak olduğu çıkarımı yapılabilirdi.

Liste2'de gözlemler 1'er 1'er artmıştır. Birbirine yakındır.

List2'nin varyansı 2'dir. Düşük varyans değeri verinin değerlerinin birbirine yakın olduğunu gösterir.

**Standard Deviation:** Veri setindeki gözlemlerin dağılımının ölçüsüdür.

Bir veri setindeki gözlemlerinin ortalamadan ne kadar farklı olduğunu ölçer.

```python
print("Standard Deviation: ", (np.std(list1)))
std = np.sqrt(sum((list1 - np.mean(list1))**2)/len(list1))
print("Standard Deviation with formula: ",std)
print("Standard Deviation: ", (np.std(list2)))
std = np.sqrt(sum((list2 - np.mean(list2))**2)/len(list2))
print("Standard Deviation with formula: ",std)
```

List1'deki değerler sırasıyla -10,0,10,20,30'dur ve ortalaması 10'dur.

Gözlemler ortalamadan uzak olduğu için standart sapması 14 gibi büyük bir değerdir.

# **Quartile**

python-outlier-detection-example-1.jpg

*  **Median** değeri listenin ortasındaki sayıdır ve second quartile (50%) olarak da adlandırılır.

*  **Lower quartile** listedeki en küçük sayı ile listenin medianının ortasındaki sayıdır, first quartile (25%) olarak da adlandırılır.

*  **Upper quartile** listenin medianı ile listedeki en büyük sayının ortasındaki sayıdır, third quartile (75%) olarak da adlandırılır.

*  Lower quartile(Q1) ile upper quartile(Q3) arasındaki mesafe **interquartile** range'dir.

*  Q1 - 1.5(Q3-Q1) değerinin altındaki değerler listedeki minimum değerin altında kalır ve outlierdır.

*  Q3 + 1.5(Q3-Q1). (Q3-Q1) değerinin üstündeki değerler listedeki maximum değerin üstünde kalır ve outlierdır.

```python
import pandas as pd

#read data
data = pd.read_csv("/content/data.csv")
data = data.drop(['Unnamed: 32','id'],axis = 1)
```

```python
data.head()
```

```python
data_bening = data[data["diagnosis"] == "B"]
data_malignant = data[data["diagnosis"] == "M"]
```

```python
data.describe()
```

Describe methoduna bakılarak lower quartile (25%), upper quartile (75%) ve median (50%) değerleri gözükür.

Ayrıca radius mean feature'ına bakıldığında ortalaması 14.12, standart sapması 3.52'dir. İlk 5 satıra da bakıldığında sample'ların ortalamaya yakın değerler olduğu görülür.

area_mean feature'ına bakıldığında ortalaması 654.88, standart sapması 351.91'dir. Buradan bu feature'daki sampleların ortalamadan uzak olduğu çıkarımı yapılır.

```python
desc = data_bening.radius_mean.describe()
Q1 = desc[4]
Q3 = desc[6]
IQR = Q3-Q1
lower_quartile = Q1 - 1.5*IQR
upper_quartile = Q3 + 1.5*IQR
print("Anything outside this range is an outlier: (", lower_quartile ,",", upper_quartile,")")
```

İyi huylu tümör için radius mean feature'ında lower_quartile ve upper_quartile değerleri hesaplanmıştır.

Bu değerler sınır değerleridir ve dışında kalan sample'lar outlier'dır.

```python
print("Outliers: ",data_bening[(data_bening.radius_mean < lower_quartile) | (data_bening.radius_mean > upper_quartile)].radius_mean.values)
```

Lower quartile ve upper quartile'ın dışında kalan 3 outlier olduğu görülür.

**Covariance:** İki variable'ın birlikte nasıl değiştiğidir.

# **Pearson Correlation Coefficient**

**Pearson Correlation Coefficient:** Covariance'ın standardize edilmiş halidir.

Covariance standard deviation'a bölünerek -1 ile 1 arasında standardize edilmiş değerler elde edilir.

Pearson Correlation Coefficient değeri;

1'e yaklaştıkça positive linear corelation

-1'e yaklaştıkça negatif linear corelation

0'a yaklaştıkça corelation olmadığı çıkarımında bulunulur.

```python
p1 = data.loc[:,["area_mean","radius_mean","fractal_dimension_se"]].corr(method= "pearson")
p2 = data.radius_mean.cov(data.area_mean)/(data.radius_mean.std()*data.area_mean.std())
p3 = data.radius_mean.cov(data.fractal_dimension_se)/(data.radius_mean.std()*data.fractal_dimension_se.std())
print('Pearson correlation: ')
print(p1)
print('Pearson correlation between area_mean and radius_mean: ',p2)
print('Pearson correlation between radius_mean and fractal_dimension_se: ',p3)
```

radius_mean ve area_mean arasındaki pearson correlation coefficient değeri 0.98'dir. 

Daha önce çizilmiş joint plot, pair plot, scatter plotta da görüldüğü gibi iki feature arasında pozitif yönlü linear güçlü bir corelation vardır.

radius_mean ve fractal_dimension_se featureları arasındaki pearson correlation coefficient değeri -0.04'tür. 0'a çok yakın bir değer olduğundan aralarında herhangi bir corelation yoktur.

# **Spearman Rank Coefficient**

İlk başta rank'i anlatmak için bir örnek yapılacaktır.

```python
data1 = {'name': ['ali', 'veli', 'hakan', 'ayse', 'fatma'], 
        'year': [2012, 2012, 2013, 2014, 2014], 
        'reports': [4, 24, 31, 2, 3],
        'coverage': [25, 94, 57, 62, 70]}
```

Bir dictionary oluşturulur.

Bu dictionary'de name, year, reports, coverage olmak üzere 4 farklı feature bulunmaktadır.

```python
df = pd.DataFrame(data1, index = ['ankara', 'istanbul', 'sinop', 'bolu', 'izmir'])
df
```

Oluşturulan dictionary şehirlere göre indekslenerek dataframe'e çevrilir.

```python
df['coverageRanked'] = df['coverage'].rank(ascending=1)
df
```

Covarage feature'ına göre artan bir şekilde **rank** yapıldığında en küçük değere sahip sample'ın rank değeri 1'dir.

5 feature olduğundan en büyük coverage değerine sahip sample'ın rank değeri 5'tir.

Artan bir şekilde **rank** yapıldığında sample'lar 1'den maximum feature sayısına kadar belirtilen feature'daki değerlerine göre bir rank değeri elde ederler.

Pearson Correlation Coefficient eğer outlierlar varsa handle edemediğinden iyi sonuç vermeyecektir. Bu yüzden robust (dayanıklı) değildir.

Ancak Spearman Rank Coefficient outlierları handle edebilen daha robust bir methoddur.

Datanın dağılımı normal bir dağılım olduğunda iki feature arasındaki ilişkiye bakmak için Pearson Correlation Coefficient kullanılırken, normal bir dağılım olmadığında Spearman Rank Coefficient kullanılır.

```python
ranked_data = data.rank() 
spearman_corr1 = ranked_data.loc[:,["area_mean","radius_mean"]].corr(method= "pearson")
spearman_corr2 = ranked_data.loc[:,["area_mean","fractal_dimension_se"]].corr(method= "pearson")
print("Spearman's correlation: ")
print(spearman_corr1)
print(spearman_corr2)
```

Spearman corelation'a bakılırken pearson methodu kullanılır.

Ancak kullanılan data rank edilmiş bir datadır.

radius_mean ve area_mean arasındaki spearman correlation coefficient değeri 0.99'dur, iki feature arasında pozitif yönlü linear güçlü bir corelation vardır.

radius_mean ve fractal_dimension_se featureları arasındaki spearman correlation coefficient değeri -0.01'tür. 0'a çok yakın bir değer olduğundan aralarında herhangi bir corelation yoktur.

radius_mean ve area_mean featureları için Spearman corelatin değeri pearson corelation değerinden daha yüksek çıkmıştır.

Bu sonuçtan datada outlierlar olduğu ve pearson'ın değerinin bu outlierlardan etkilendiği çıkarımı yapılır.

# **Effect size**

Effect size istatistik kavramlarından bir tanesidir.

İki farklı grup arasındaki benzerliği ele alır, iki grup arasındaki farklılığı vurgular.

Cohen effect size'a göre; 

effect size (d) değeri;

*  0.2 ise iki grubun birbirine benzediği,

*  0.5 ise benzerlik bulunduğu ancak çok fazla olmadığı,

*  0.8 ise iki grubun birbirine benzemediği çıkarımı yapılır.

**effect size formülü**

Formula-for-Cohens-d.png

**M:** mean değeri

**SD:** standard deviation değeri

standart deviation'ın karesi alındığında variance değeri ortaya çıkacağından iki grubun variance'ına bakılır. 

Tümör datası iyi huylu ve kötü huylu olmak üzere quartile değerine bakılırken iki gruba ayrılmıştı.

İyi huylu kanser hücrelerinin yarıçaplarından bir grup, kötü huylu kanser hücrelerinin yarıçaplarından başka bir grup oluşturularak iki grup karşılaştırılacaktır.

```python
#m2-m1
mean_diff = data_malignant.radius_mean.mean() - data_bening.radius_mean.mean()
```

```python
var_bening = data_bening.radius_mean.var() #std1^2
var_malignant = data_malignant.radius_mean.var() #std2^2
```

```python
var_pooled = (len(data_bening)*var_malignant +len(data_malignant)*var_bening ) / float(len(data_bening)*len(data_malignant))
```

```python
effect_size = mean_diff/np.sqrt(var_pooled)
print("Effect size: ",effect_size)
```

Effect size değeri 22 çıkmıştır.

Bu çok büyük bir değer olduğundan iki grup arasında bir benzerlik yoktur, birbirinden farklıdır.

Zaten iki grubun birisi iyi huylu kanser hücresi, diğeri kötü huylu kanser hücresi olduğundan birbirinden farklı oldukları bilinmektedir.
