# **Chi Square Analysis**

Gözlenen ve beklenen frekanslar arasındaki farkın anlamlı olup olmadığı temeline dayanır.

image.png

Bir örnekle anlatılacak olursa;

Havaya 10 kere para atıldığında 9 kere yazı 1 tura gelmiştir.

Null hypothesis: 10 atıştan 9 kere yazı gelmesi istatiksel olarak mantıklıdır.

Beklene sonuç %50-%50 eşit çıkmasıdır.

```python
observation_tura = 1
expected_tura = 5
```

```python
observation_yazı = 9
expected_yazı = 5
```

```python
import numpy as np
chi_value = np.sum((((observation_tura-expected_tura)**2)/expected_tura) + (((observation_yazı-expected_yazı)**2)/expected_yazı))
chi_value
```

Formüle göre chi square value 6.4'tür.

image.png

Chi square table'a bakıldığında,

degrees of freedom değeri, olası sonuçların sayısı - 1 ile bulunur.

Örnekte yazı ve tura olmak üzere 2 sonuç vardır.

2-1 = 1

confidence level 0.05 seçildiğinde, null hypothesis 100 kez test edilip 95'inde reject edilmiştir.

Örnek için bu değer 3.84'tür.

6.4 > 3.84 olduğu için null hypothesis reject edilmiştir.

Ayrıca yapılan araştırmalara göre madeni paranın tura tarafı, yazı tarafından daha ağır olduğu için para havaya atıldığında yazı gelme ihtimali daha fazladır.
